```python
import numpy as np
import tweepy
import json
import pandas as pd
import matplotlib.pyplot as plt
import time
import seaborn as sns

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
sentiment_array = []
sentiment_array_avg = []
```

```python
Sample = "Resources/mycsv1.csv"
Sample_df = pd.read_csv(Sample, encoding="ISO-8859-1")
Sample_df.head()
```

```python
text1 = Sample_df['comment_text']
print(text1)
```

```python
 # Variables for holding sentiments
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
results = []
```

```python
# Run Vader Sentiment Analysis on Each of the comments
# Loop through Each Sample
for text in text1:

    # Run Vader Analysis on each Sample
    results = analyzer.polarity_scores(text)
    compound = results["compound"]
    pos = results["pos"]
    neu = results["neu"]
    neg = results["neg"]

    # Print Samples and Analysis
    print(text)
    print("Compound Score:", compound)
    print("Positive Score:", pos)
    print("Neutral Score:", neu)
    print("Negative Score: ", neg)
    print()
```

```python
    # Storing the Average Sentiments
sentiments = {"Compound Score": compound_list, "Positive Score": positive_list, "Negative Score": negative_list, "Neutral Score": neutral_list,}
    
sentiment_array_avg.append({ "Compound Score": np.average(compound_list), "Positive Score": np.average(positive_list), "Negative Score": np.average(negative_list), "Neutral Score": np.average(neutral_list)})
    
sentiment_array.append(sentiments)
print(sentiments)
```
